```python
plt.ylim(-2, 2)
plt.xticks([-numpy.pi, -numpy.pi/2, 0, numpy.pi/2, numpy.pi], 
           [r'$-\pi$', r'$-\pi/2$', '0', r'$\pi/2$', r'$\pi$'])
plt.title('The sine function')
                                                            
plt.plot(X, Y, )
```

```python
plt.ylim(-2, 2)
plt.xticks([-numpy.pi, -numpy.pi/2, 0, numpy.pi/2, numpy.pi], 
           [r'$-\pi$', r'$-\pi/2$', '0', r'$\pi/2$', r'$\pi$'])
plt.title('The sine function')
X1 = X
X2 = X + numpy.pi/2
plt.plot(X1, Y, X2, Y)
```

```python
plt.ylim(-2, 2)
plt.xticks([-numpy.pi, -numpy.pi/2, 0, numpy.pi/2, numpy.pi], 
           [r'$-\pi$', r'$-\pi/2$', '0', r'$\pi/2$', r'$\pi$'])
plt.title('The sine function')
X1 = X
X2 = X + numpy.pi/2
plt.plot(X1, Y, label='sine')
plt.plot(X2, Y, label='cosine')
plt.legend(loc='upper left')
```
